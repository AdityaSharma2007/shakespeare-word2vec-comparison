```python
import os
import nltk
nltk.download('all')
```

```python
!pip install gensim
```

```python
print(os.getcwd())
print(os.listdir())
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

story = []



for filename in os.listdir():

    if filename.endswith(".txt"):

        with open(filename, 'r', encoding='latin-1') as f:
            text = f.read()
            sentences = sent_tokenize(text)

            for sentence in sentences:
                tokens = simple_preprocess(sentence)
                story.append(tokens)

print("Total sentences:", len(story))
print(story)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))  # define once

clean_story = []

for sen in story:
    filter = [word for word in sen if word not in stop_words]
    clean_story.append(filter)

story = clean_story

story
```

```python
from gensim.models import Word2Vec

model_cbow= Word2Vec(
    sentences=story,
    vector_size=200,
    window=10,
    min_count=2,
    workers=4,
    epochs=10,
    sg=0   # CBOW
)

model_skipgram = Word2Vec(
    sentences=story,
    vector_size=200,
    window=10,
    min_count=2,
    workers=4,
    epochs=10,
    sg=1   # Skip-gram
)
```

```python
print("\nMost similar to 'king':\n")
print(model_cbow.wv.most_similar('king'))

print("\nMost similar to 'love':\n")
print(model_cbow.wv.most_similar('love'))

print("\nMost similar to 'death':\n")
print(model_cbow.wv.most_similar('death'))

print("\nSimilarity between 'king' and 'queen':\n")
print(model_cbow.wv.similarity('king', 'queen'))
```

```python
print("\nMost similar to 'king':\n")
print(model_skipgram.wv.most_similar('king'))

print("\nMost similar to 'love':\n")
print(model_skipgram.wv.most_similar('love'))

print("\nMost similar to 'death':\n")
print(model_skipgram.wv.most_similar('death'))

print("\nSimilarity between 'king' and 'queen':\n")
print(model_skipgram.wv.similarity('king', 'queen'))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
x1 = pca.fit_transform(model_cbow.wv.get_normed_vectors())
x2 = pca.fit_transform(model_skipgram.wv.get_normed_vectors() )

y1 = model_cbow.wv.index_to_key
y2 = model_skipgram.wv.index_to_key

import plotly.express as px

fig1 = px.scatter_3d(x1[:100], x=0, y=1, z=2, color=y1[:100] )
fig2 = px.scatter_3d(x2[:100], x=0, y=1, z=2, color=y2[:100] )

fig1.show()
fig2.show()
```
